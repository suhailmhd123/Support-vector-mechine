```python
# Split the x and y
x = data.iloc[:,:12]
y = data["DEATH_EVENT"]
```

```python
# Split the train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Model Building

```python
# Apply SVM
from sklearn import svm
clf = svm.SVC(kernel='rbf')
```

```python
clf.fit(x_train, y_train)
```

```python
# Prediction on train
y_pred_train = clf.predict(x_train)
# Prediction on test
y_pred_test = clf.predict(x_test)
```

```python
# Accuracy score for train and test
print("Accuracy score for train:" , clf.score(x_train, y_train))
print("Accuracy score for test:" , clf.score(x_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# RMSE
mse = mean_squared_error(y_train, y_train_r_pred)
train_rmse = np.sqrt(mse)
print("Train_RMSE:" ,train_rmse)

mse = mean_squared_error(y_test, y_test_r_pred)
test_rmse = np.sqrt(mse)
print("Test_RMSE:" ,test_rmse)
```

```python
# R-Squared
# For train
train_rsquared = r2_score(y_train, y_train_r_pred)
print("R-squared train data:" ,train_rsquared)

# For test
test_rsquared = r2_score(y_test, y_test_r_pred)
print("R-squared test data:" ,test_rsquared)
```

```python
# Check to the Adjusted R Squared
def adjusted_r_squared(r_squared, n, k):
    adjusted_r2 = 1 - (1 - r_squared)*(n - 1)/(n - k - 1)
    return adjusted_r2
```

```python
# For train data
n_train, k = x_train.shape[0], x_train.shape[1]
train_adjusted_r = adjusted_r_squared(train_rsquared, n_train, k)
# For test data
n_test = x_test.shape[0]
test_adjusted_r = adjusted_r_squared(test_rsquared, n_test, k)
```

```python
print("Adjusted R_Squared for train:" ,train_adjusted_r)
print("Adjusted R_Squared for test:" ,test_adjusted_r)
```
